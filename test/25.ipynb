```sql
SELECT * FROM pop_area_2010 ORDER BY pop_2010 DESC LIMIT 3;
```

```sql
SELECT * FROM pop_area_2010 WHERE pop_2010 < 5000;
```

```sql
SELECT a.city, pop_2010, pop_2020
FROM cities AS a, pop_area_2010 AS b 
WHERE a.city = b.city AND a.county = b.county AND a.state = b.state;
```

```sql
SELECT a.city, b.city
FROM cities AS a, cities AS b
WHERE a.county = b.county AND a.city < b.city;
```

```sql
SELECT (city || ", " || state) AS full_city, (county || " County") AS full_county
FROM cities
LIMIT 5;
```

```sql
SELECT pop_2020 / 1000 AS pop_1000s FROM cities;
```

```sql
SELECT SUBSTR(city, 1, 3) AS initials
FROM cities;
```

```sql
SELECT a.city, a.pop_2020 - b.pop_2010 AS pop_difference
FROM cities AS a, pop_area_2010 AS b
WHERE a.city = b.city AND a.county = b.county AND a.state = b.state ORDER BY pop_difference DESC LIMIT 1;
```

```sql
SELECT pop_2020
FROM cities
ORDER BY pop_2020 DESC
LIMIT 1;
```

```sql
SELECT MAX(pop_2020)
FROM cities;
```
